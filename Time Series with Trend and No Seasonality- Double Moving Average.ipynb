```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('MA.csv')
df.head()
```

```python
df.tail()
```

```python
df['3_month_MA_1'] = 0.00
for i in range(0,38):
    df['3_month_MA_1'][i+3] = np.mean( [ df['Y_true'][i], df['Y_true'][i+1], df['Y_true'][i+2] ])
df.head(6)                                
```

```python
df.tail()
```

```python
df['3_month_MA_2'] = 0.00
for i in range(3,38):
    df['3_month_MA_2'][i+2] = np.mean( [ df['3_month_MA_1'][i], df['3_month_MA_1'][i+1], df['3_month_MA_1'][i+2] ])
df.head(8)                                
```

```python
df.tail()
```

```python
df['Y_pred'] = 0.00
for i in range(3,38):
    df['Y_pred'][i+2] = 2 * df['3_month_MA_1'][i+2] - df['3_month_MA_2'][i+2] + (2/(3-1))*(df['3_month_MA_1'][i+2] - df['3_month_MA_2'][i+2])
df.head(8)       
```

```python
df.tail()  
```

```python
df['Absolute Error'] = 0.00
for i in range(5,39):
    df['Absolute Error'][i] = abs( df['Y_pred'][i] - df['Y_true'][i])
df.head(8)                                
```

```python
df.tail()
```

```python
df['Squared Error'] = 0.00
for i in range(5,39):
    df['Squared Error'][i] = (df['Y_pred'][i] - df['Y_true'][i])**2
df.head(8)                                
```

```python
df.tail()
```

```python
df['Precent Error1'] = 0.00
for i in range(5,39):
    df['Precent Error1'][i] = abs(( df['Y_true'][i] - df['Y_pred'][i] ) / df['Y_true'][i])
df.head(8)                                
```

```python
df.tail()
```

```python
df['Precent Error2'] = 0.00
for i in range(5,39):
    df['Precent Error2'][i] = ((df['Y_pred'][i] - df['Y_true'][i])/ df['Y_true'][i-1])**2
df.head(8)                                
```

```python
df.tail()
```

```python
df['Precent Error3'] = 0.00
for i in range(5,39):
    df['Precent Error3'][i] = ((df['Y_true'][i] - df['Y_true'][i-1] ) / df['Y_true'][i-1] )**2
df.head(8)                                
```

```python
df.tail()
```

```python
df['Error'] = 0.00
for i in range(5,39):
    df['Error'][i] = ( df['Y_pred'][i] - df['Y_true'][i] )
df.head(8)                                
```

```python
df.tail()
```

```python
df['Squared Error Diff'] = 0.00
for i in range(6,39):
    df['Squared Error Diff'][i] = "{:,.2f}".format((df['Error'][i] - df['Error'][i-1] )**2)
df.head(9)                                
```

```python
df.tail()
```

```python
MSE = np.mean(df['Squared Error'][5:-1])
MSE
```

```python
RMSE = np.sqrt(np.mean(df['Squared Error'][5:-1]))
RMSE
```

```python
MAD = np.mean(df['Absolute Error'][5:-1])
MAD
```

```python
MAPE = np.mean(df['Precent Error1'][5:-1])
MAPE
```

```python
ThielsU = np.sqrt( np.sum( df['Precent Error2'][5:-1] ) / np.sum( df['Precent Error3'][5:-1] ) )
ThielsU 
```
